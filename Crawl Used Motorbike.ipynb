```python
# Import libraries and packages for the project 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
import csv
import requests
```

```python
#  Open Chrome and Access Linkedin login site
driver = webdriver.Chrome(ChromeDriverManager().install())
sleep(2)
url = 'https://xe.chotot.com/mua-ban-xe-may'
driver.get(url)
sleep(2) 
```

```python
#  Scrape the URLs of the profiles
#  Write a function to extract the URLs of one page
def GetURL():
    page_source = BeautifulSoup(driver.page_source)
    profiles = page_source.find_all('a', class_ = 'AdItem_adItem__gDDQT') 
    all_profile_URL = []
    for profile in profiles:
        profile_ID = profile.get('href')
        profile_URL = "https://xe.chotot.com/mua-ban-oto" + profile_ID
        profile_URL = profile.get('href')
        if profile_URL not in all_profile_URL:
            all_profile_URL.append(profile_URL)
    return all_profile_URL
```

```python
#  Navigate through many page, and extract the profile URLs of each page
input_page = int(input('How many pages you want to scrape: '))
URLs_all_page = []

page_source = BeautifulSoup(driver.page_source, "html.parser")
p=498
for page in range(input_page):
    p+=1
    url = 'https://xe.chotot.com/mua-ban-xe-may' + '?page=' + str(p) 
    driver.get(url)
    URLs_one_page = GetURL() 
    sleep(2)
    URLs_all_page = URLs_all_page + URLs_one_page
    sleep(2) 
```

```python
#  Scrape the data of 1 Linkedin profile, and write the data to a .CSV file
from pyrsistent import thaw
c=['Title', 'Price', 'Brand', 'Model', 'Reg_Date', 'Km',
            'Condition','Capacity', 'Type', 'Location']
baseUrl = "https://xe.chotot.com/"
df=pd.DataFrame(columns=c)
s=0

for linkedin_URL in URLs_all_page:
    driver.get(baseUrl + linkedin_URL)
    print('- Accessing profile: ', linkedin_URL)
    s+=1
    sleep(1)
    page_source = BeautifulSoup(driver.page_source, "html.parser")
    ft=[]
    try:
        name = page_source.find('h1', itemprop="name").get_text()
    except:
        name = ""
    ft.append(name)
        
    try:
        price = page_source.find('span', itemprop="price").get_text()
    except:
        price = ""
    ft.append(price)
        
    try:
        brand = page_source.find('a', itemprop="motorbikebrand").get_text()
    except:
        brand = ""
    ft.append(brand)
        
    try:
        model = page_source.find('a', itemprop="motorbikemodel").get_text()
    except:
        model = ""
    ft.append(model)
        
    try:
        mfdate = page_source.find('span', itemprop="regdate").get_text()
    except:
        mfdate = ""
    ft.append(mfdate)
    
    try:
        mileage_v2 = page_source.find('span', itemprop="mileage_v2").get_text()
    except:
        mileage_v2 = ""
    ft.append(mileage_v2)
    
    try:
        condition_ad = page_source.find('span', itemprop="condition_ad").get_text()
    except:
        condition_ad = ""
    ft.append(condition_ad)    
          
    try:
        fuel = page_source.find('span', itemprop="motorbikecapacity").get_text()
    except:
        fuel = ""
    ft.append(fuel)

    try:
        cartype = page_source.find('span', itemprop="motorbiketype").get_text()
    except:
        cartype = ""
    ft.append(cartype)    
    
    try:
        fz1 = page_source.find('span', class_="fz13").get_text()
    except:
        fz1 = ""
    ft.append(fz1)
    
    df.loc[len(df)] = ft
```
